```python
def normalization(X):
    
    standard_diviation = np.std(X , axis = 0)
    
    mean = np.mean(X, axis = 0)
    
    X_normalized = (X - mean) / standard_diviation
    
    return X_normalized
```

```python
X = normalization(X)
```

```python
X = np.concatenate([np.ones(m).reshape(m , 1) , X] , axis  = 1)
y = y.reshape(m , )
```

```python
print('The shape of the X is {}'.format(X.shape))
print('The shape of the y is {}'.format(y.shape))
```

```python
def cost_function(X , y , theta):
    
    
    m = X.size
    
    J = 0
    
    h = np.dot(X , theta)
    
    J = 1 / (2 * m) * (np.sum(np.square(h - y))) 
    
    return J
```

```python
theta = np.zeros(8)
J = cost_function(X , y , theta)
```

```python
print('The value of the cost function with these {} theta is {:.2f} .'.format(theta , J))
```

```python
def gradian_decent(X , y , theta , alpha , iteration):
    
    m = X.size 
    
    theta = theta.copy()
    
    Jz = []
    
    thetaz = []
    
    for i in range(iteration):
        
        theta = theta - ( alpha / m ) *  ( np.dot( X , theta ) - y ).dot( X )
        
        thetaz.append(theta)
        
        Jz.append(cost_function(X , y , theta))
        
        
    minimized_J = Jz[-1]
    
    best_thetas = thetaz[-1]
    
    return minimized_J , best_thetas , Jz
```

```python
theta = np.zeros(8)
alpha = .02
iteration = 2000
```

```python
minimized_J , best_thetas , Jz = gradian_decent(X , y , theta , alpha , iteration)
```

```python
print("Theta found by gradient descent: {}".format(best_thetas))
print("Final cost value:  {:0.10f}".format(minimized_J))
```
